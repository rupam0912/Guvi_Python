```
#Print all the value in the list
#print the hex value of green
#print the hex codes of all colors

l = [
    {
        "color": "red",
        "value": "#f00"
    },
    {
        "color": "green",
        "value": "#0f0"
    },
    {
        "color": "blue",
        "value": "#00f"
    }
]

print("Values:")
for i in range(len(l)):
  print(l[i]["value"],end ="\n")

print()
print("hex value of green->")
for i in range(len(l)):
  if l[i]["color"] == "green":
    print(l[i]["value"])
print()


print("hex code of all the colors->")
for i in range(len(l)):
  print(l[i]["color"],":",l[i]["value"])
```

```
wtf = {2: [1, 2, 3], 'a': {'b': {'c': {'d': {'e': [1, 2, 3]}}}}}
#Get the first array value for the key 2
print("first value: ", wtf[2][0])
#Print all the array value for the key 2
print("All value: ", wtf[2])
#Print value of key 'a','b','c','d','e'
print("a: ",wtf["a"] )
print("b: ",wtf["a"]["b"] )
print("c: ",wtf["a"]["b"]["c"] )
print("d: ",wtf["a"]["b"]["c"]["d"] )
print("e: ",wtf["a"]["b"]["c"]["d"]["e"] )
#Print the sum of the array with key 'e'
print("sum: ", sum(wtf["a"]["b"]["c"]["d"]["e"]))
#set value of key 'e' to ['Chera','Chola','Pandiya']
wtf["a"]["b"]["c"]["d"]["e"] = ['Chera','Chola','Pandiya']
print("modified e: ",wtf["a"]["b"]["c"]["d"]["e"] )
```
